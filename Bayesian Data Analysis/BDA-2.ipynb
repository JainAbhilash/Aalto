```r
#1) We first read the data:
data=readLines("algae.txt")
data=as.integer(data)
x=seq(0,1,length=10000)
prior = dbeta(x,2,10)
options(repr.plot.res=300)
#Plot the prior distribution
plot(x,prior,'l')
n=length(data) #Total number of observations 
y=sum(data) #Number of observations with value=1
cat('n =',n)#274
cat('\ny =',y)#44
```

```r
alpha=2
beta=10
#Plotting the posterior Distribution
posterior = dbeta(x,alpha+y,beta+(n-y))
plot(x,posterior,'l')
```

```r
prior_mean= alpha/(alpha+beta)
posterior_mean=(alpha+y)/((alpha+y)+(beta+n-y))
#Point estimate is the mean of the posterior distribution i.e 0.160839161
cat('point estimate = ',posterior_mean)
#Interval estimates are the 95% central confidence intervals i.e the 0.025 and the 0.975 interval=[0.12065,0.205512]
posterior_conf_int=c(qbeta(0.025,alpha+y,beta+(n-y)),qbeta(0.975,alpha+y,beta+(n-y)))
cat('\ninterval estimate= [',posterior_conf_int,']')
```

### 4) We try prior sensitivity by taking 1) Uniform Distribution(alpha=1,beta=1) and were the values of alpha is higher than that of beta(alpha=15 and beta =1)

```r
#uniform distribution
n_alpha=1
n_beta=1
n_prior=dbeta(x,n_alpha,n_beta)
plot(x,n_prior,'l')
```

```r
#Posterior Distribution after Uniform Prior
n_posterior=dbeta(x,n_alpha+y,n_beta+(n-y))
plot(x,n_posterior,'l')
```

```r
#Now we try to estimate the difference between the posterior we obtain from the question and our tested posterior
posterior_mean=(n_alpha+y)/((n_alpha+y)+(n_beta+n-y))
cat('New posterior mean = ',posterior_mean)

posterior_conf_int=c(qbeta(0.025,n_alpha+y,beta+(n-y)),qbeta(0.975,n_alpha+y,n_beta+(n-y)))
cat('\nNew posterior confidence interval = [',posterior_conf_int,']')
```

### Conclusion: We see a slight difference between the point estimates and the interval estimates,meaning that our posterior is not that sensitive
